```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nootropiche = ["Ashwagandha", "Bacopa", "Bobinsana", "Efedra", "Fagiolo Vellutato", "Ginkgo Biloba", "Ginseng", "Ilex Guayasa", "Moringa", "Rodiola Rosea", "Té nero", "Té verde", "Yerba Mate", ]
sedative = ["Akuamma", "Artemisia", "Biancospino", "Camomilla", "Coda di Leone", "Corydalis", "Damiana", "Gelsomino", "Lattuga Selvatica", "Papavero della california", "Papavero Selvatico", "Passiflora", "Pedicularis Hirsuta", "Kava", "Loto Blu", "Loto Rosa", "Marihuanilla", "Melissa", "Mulungu", "Skullcap", "Valeriana"]
oneirogene = ["Calea Zacatechichi", "Silene Capensis"]
salutari = ["Calendula", "Edera", "Iperico", "Spirulina", "Pau D'Arco", "Verbasco"]
ritualistiche = ["Sanango", "Sinicuichi"]
aromatiche = ["Bucce Arancia", "Galanga", "Lavanda", "Menta piperita", "Rosa di Damasco"]
tipi = ["nootropiche", "sedative", "salutari", "aromatiche", "ritualistiche", "oneirogene"]
piante = nootropiche + sedative + salutari + aromatiche + ritualistiche + oneirogene

columns = ["Nome", "Tipo", "Voto_Effetto", "Voto_Sapore", "Caratteristiche Principali", "Forma", "Concentrazione", "Dosaggio", "Modalita_Assunzione", "Descrizione", "Link", "Prezzo"]
```

```python
tipi_piante = list()
for tipo in tipi:
    for pianta in range(len(list(globals()[tipo]))):
        tipi_piante.append(tipo)
```

```python
for column in columns[2:4]:
    df[column] = 0
for column in columns[4:]:
    df[column] = None
    
df.to_csv('Piante.csv', index=False)
df
```

```python
mix = {}
data = list()
for tipo in tipi:    
    var_tipo = globals()[tipo]
    mix[tipo] = {}
    for n in range(2,4):
        mix[tipo][n] = list()
        done = list()
        if (n == 2):
            for pianta1 in var_tipo:
                done.append(pianta1)
                diff = list(set(var_tipo) - set(done))
                diff.sort()
                #print(mix[tipo])
                #print("Rimasti:")
                #print(diff)
                for pianta2 in diff:
                    mix_l = pianta1 + ', ' + pianta2
                    mix[tipo][2].append(mix_l)
                    data.append(mix_l)
        if (n == 3):
            mix_fatti = list()
            for mix2 in mix[tipo][2]:
                fatti = mix2.split(", ")
                diff = list(set(var_tipo) - set(fatti))
                diff.sort()
                #print("Rimasti mix:")
                #print(diff)
                for pianta3 in diff:
                    #print("Pianta3:", pianta3)
                    insert = 1
                    pianta1 = fatti[0]
                    pianta2 = fatti[1]
                    mix_l = mix2 + ', ' + pianta3
                    
                    for mix_fatto in mix_fatti:
                        #print("Mix fatto: ",mix_fatto)
                        if (pianta1 in mix_fatto and pianta2 in mix_fatto and pianta3 in mix_fatto):
                            insert = 0
                    if (insert == 1):
                        mix[tipo][3].append(mix_l)
                    str_mix_fatti = mix_l.split(", ")
                    mix_fatti.append(str_mix_fatti)
                    data.append(mix_l)
mix
```

```python
mix_aromatiche = {}
for tipo in ["nootropiche", "sedative", "salutari"]:    
    var_tipo = globals()[tipo]
    print("Var tipo: ",var_tipo)
    mix_aromatiche[tipo] = {}
    for n in range(2,4):
        mix_aromatiche[tipo][n] = list()
        done = list()
        if (n == 2):
            for pianta1 in var_tipo:
                done.append(pianta1)
                diff = list(set(var_tipo) - set(done))
                diff.sort()
                #print(mix[tipo])
                #print("Rimasti:")
                #print(diff)
                for pianta2 in aromatiche:
                    mix_l = pianta1 + ', ' + pianta2
                    mix_aromatiche[tipo][2].append(mix_l)
                    data.append(mix_l)
        if (n == 3):
            mix_fatti = list()
            for mix2 in mix_aromatiche[tipo][2]:
                fatti = mix2.split(", ")
                diff = list((set(var_tipo).union(set(aromatiche))) - set(fatti))
                diff.sort()
                #print("Rimasti mix:")
                #print(diff)
                for pianta3 in diff:
                    #print("Pianta3:", pianta3)
                    insert = 1
                    pianta1 = fatti[0]
                    pianta2 = fatti[1]
                    mix_l = mix2 + ', ' + pianta3
                    
                    for mix_fatto in mix_fatti:
                        #print("Mix fatto: ",mix_fatto)
                        if (pianta1 in mix_fatto and pianta2 in mix_fatto and pianta3 in mix_fatto):
                            insert = 0
                    if (insert == 1):
                        mix_aromatiche[tipo][3].append(mix_l)
                        str_mix_fatti = mix_l.split(", ")
                        mix_fatti.append(str_mix_fatti)
                        data.append(mix_l)
mix_aromatiche
```

```python
all_mix = list()
for tipo in mix:
    for num in mix[tipo]:
        for m in mix[tipo][num]:
            all_mix.append(m)

for tipo in mix_aromatiche:
    for num in mix_aromatiche[tipo]:
        for m in mix_aromatiche[tipo][num]:
            all_mix.append(m)
            
all_mix
```

```python
columns = ["Voto_Effetto", "Voto_Sapore", "Caratteristiche Principali"]

df = pd.DataFrame(all_mix, columns=["Nome"])

for column in columns[0:2]:
    df[column] = 0
df[columns[2]] = None

df.to_csv('Mix_with_aromatiche.csv', index=False)

df
```
